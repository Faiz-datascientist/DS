1. Reverse a string without using any built-in string reversal functions:
```python
def reverse_string(s):
    reversed_str = ""
    for i in range(len(s)-1, -1, -1):
        reversed_str += s[i]
    return reversed_str

# Example usage
string = "Hello, World!"
reversed_string = reverse_string(string)
print(reversed_string)
```

2. Check if a given string is a palindrome:
```python
def is_palindrome(s):
    s = s.lower()  # Convert string to lowercase for case-insensitive comparison
    reversed_str = reverse_string(s)
    return s == reversed_str

# Example usage
string = "Madam"
if is_palindrome(string):
    print("It is a palindrome.")
else:
    print("It is not a palindrome.")
```

3. Find the largest element in a given list:
```python
def find_largest_element(lst):
    largest = lst[0]
    for num in lst:
        if num > largest:
            largest = num
    return largest

# Example usage
numbers = [5, 9, 3, 7, 11, 2]
largest_number = find_largest_element(numbers)
print(largest_number)
```

4. Count the occurrence of each element in a list:
```python
def count_occurrences(lst):
    occurrence_count = {}
    for element in lst:
        if element in occurrence_count:
            occurrence_count[element] += 1
        else:
            occurrence_count[element] = 1
    return occurrence_count

# Example usage
numbers = [1, 2, 3, 2, 1, 3, 2, 4, 5, 4, 2]
occurrences = count_occurrences(numbers)
print(occurrences)
```

5. Find the second largest number in a list:
```python
def find_second_largest(lst):
    largest = float('-inf')
    second_largest = float('-inf')
    for num in lst:
        if num > largest:
            second_largest = largest
            largest = num
        elif num > second_largest and num < largest:
            second_largest = num
    return second_largest

# Example usage
numbers = [5, 9, 3, 7, 11, 2]
second_largest_number = find_second_largest(numbers)
print(second_largest_number)
```

```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Example usage
number = 5
factorial_result = factorial(number)
print(factorial_result)
```

8. Check if a given number is prime:
```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example usage
number = 17
if is_prime(number):
    print("It is a prime number.")
else:
    print("It is not a prime number.")
```

13. Generate all permutations of a given string:
```python
def generate_permutations(s):
    if len(s) <= 1:
        return [s]
    permutations = []
    for i in range(len(s)):
        char = s[i]
        remaining = s[:i] + s[i+1:]
        for perm in generate_permutations(remaining):
            permutations.append(char + perm)
    return permutations

# Example usage
string = "abc"
permutations = generate_permutations(string)
print(permutations)
```

14. Calculate the Fibonacci sequence up to a given number of terms:
```python
def fibonacci(n):
    sequence = [0, 1]
    for i in range(2, n):
        sequence.append(sequence[i-1] + sequence[i-2])
    return sequence

# Example usage
num_terms = 10
fib_sequence = fibonacci(num_terms)
print(fib_sequence)
```

15. Find the median of a list of numbers:
```python
def find_median(numbers):
    numbers.sort()
    n = len(numbers)
    if n % 2 == 0:
        mid = n // 2
        median = (numbers[mid-1] + numbers[mid]) / 2
    else:
        mid = n // 2
        median = numbers[mid]
    return median

# Example usage
numbers = [4, 2, 1, 5, 3]
median = find_median(numbers)
print(median)
```

18. Find the maximum subarray sum in a given list:
```python
def find_max_subarray_sum(lst):
    max_sum = float('-inf')
    current_sum = 0
    for num in lst:
        current_sum = max(num, current_sum + num)
        max_sum = max(max_sum, current_sum)
    return max_sum

# Example usage
numbers = [1, -2, 3, 10, -4, 7, 2, -5]
max_sum = find_max_subarray_sum(numbers)
print(max_sum)
```

22. Find the first non-repeating character in a string:
```python
def find_first_non_repeating_char(s):
    char_count = {}
    for char in s:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    for char in s:
        if char_count[char] == 1:
            return char
    return None

# Example usage
string = "abracadabra"
first_non_repeating_char = find_first_non_repeating_char(string)
print(first_non_repeating_char)
```

23. Find the prime factors of a given number:
```python
def find_prime_factors(n):
    factors = []
    i = 2
    while i * i <= n:
        if n % i:
            i += 1
        else:
            n //= i
            factors.append(i)
    if n > 1:
        factors.append(n)
    return factors

# Example usage
number = 48
prime_factors = find_prime_factors(number)
print(prime_factors)
```

24. Check if a given number is a power of two:
```python
def is_power_of_two(n):
    if n <= 0:
        return False
    return (n & (n - 1)) == 0

# Example usage
number = 16
if is_power_of_two(number):
    print("It is a power of two.")
else:
    print("It is not a power of two.")
```

25. Merge two sorted lists into a single sorted list:
```python
def merge_sorted_lists(list1, list2):
    merged_list = []
    i = 0
    j = 0
    while i < len(list1) and j < len(list2):
        if list1[i] <= list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1
    return merged_list

# Example usage
list1 = [1, 3, 5, 7]
list2 = [2, 4, 6, 8]
merged_list = merge_sorted_lists(list1, list2)
print(merged_list)
```

26. Find the mode of a list of numbers:
```python
from collections import Counter

def find_mode(numbers):
    counter = Counter(numbers)
    max_count = max(counter.values())
    mode = [num for num, count in counter.items() if count == max_count]
    return mode

# Example usage
numbers = [1, 2, 3, 2, 1, 3, 2, 4, 5, 4, 2]
mode = find_mode(numbers)
print(mode)
```

27. Find the greatest common divisor (GCD) of two numbers:
```python
def find_gcd(a, b):
    while b:
        a, b = b, a % b
    return a

# Example usage
num1 = 36
num2 = 48
gcd = find_gcd(num1, num2)
print(gcd)
```

28. Calculate the square root of a given number:
```python
def square_root(n):
    if n < 0:
        return None
    if n == 0:
        return 0
    guess = n
    while True:
        new_guess = (guess + n / guess) / 2
        if abs(guess - new_guess) < 1e-6:
            return new_guess
        guess = new_guess

# Example usage
number = 25
sqrt = square_root(number)
print(sqrt)
```

29. Check if a given string is a valid palindrome ignoring non-alphanumeric characters:
```python
import re

def is_valid_palindrome(s):
    s = re.sub(r'[^a-zA-Z0-9]', '', s).lower()
    return s == s[::-1]

# Example usage
string = "A man, a plan, a canal: Panama"
if is_valid_palindrome(string):
    print("It is a valid palindrome.")
else:
    print("It is not a valid palindrome.")
```

30. Find the minimum element in a rotated sorted list:
```python
def find_minimum(lst):
    left = 0
    right = len(lst) - 1
    while left < right:
        mid = left + (right - left) // 2
        if lst[mid] > lst[right]:
            left = mid + 1
        else:
            right = mid
    return lst[left]

# Example usage
numbers = [4, 5, 6, 7, 0, 1, 2]
minimum = find_minimum(numbers)
print(minimum)
```
